## Imports

```python
import matplotlib.pyplot as plt
import numpy as np
import pycounting as pyc
import pycsimulation as pys
```

```python
%matplotlib inline 

wlength = 100e3

# Create instances
fctrace  = pyc.Trace('data/*.count', datatype='int')
signal   = pyc.Signal()
detector = pyc.Detector()


# Determin start parameters
detector.system, fit, histogram = pyc.fit_levels(data=fctrace[0:wlength], start_parameters=(0.05, 134900, 20, 0.15, 135100, 20))
times = pyc.TimeDict(states=[0,1])

# Plot start stuff
fig = plt.figure(figsize=(20, 4))
plt.subplot2grid((1,4), (0, 0), colspan=3)
fctrace.plot(0, 10e3, c='blue')
detector.plot(color='red')
yl = plt.ylim()

plt.subplot2grid((1,4), (0, 3))
histogram.plot(inverted=True, label='Histogram')
fit.plot(histogram.bins, inverted=True, label='Fit')
detector.plot(color='red')
plt.ylim(*yl)
fig.tight_layout()

with open('signal/signal.fcs', 'w') as fobj:
    for window in fctrace.windows(wlength, nr=100):
        
        # Fit time windowth
        detector.system, fit, histogram = pyc.fit_levels(window, fit.parameters)
                
        # Digitize data
        detector.digitize(window, signal)
        
        times.add(signal)
        
        # Flush everthing
        signal.flush(fobj)

fig, ax = plt.subplots(figsize=(20, 10))
times.plot(ax, marker='o', linestyle='-', markersize=2, normed=False, log=False)
plt.xlim(0, 500);
```
