```python
%matplotlib inline 

# Create instances
fctrace  = pyc.Trace('data/*.count', datatype='int')

# Determin start parameters
wlengths = [1e4, 1e5, 1e6, 1e7]
times = pyc.MultiTime(pyc.ptime([0,1], factor=len(detectors)))

fig_t, ax_t = plt.subplots(figsize=(20, 5))
fig_t.suptitle('Window length')

for wlength, time in zip(wlengths, times):
    
    signal   = pyc.Signal()
    detector = pyc.Detector()
    
    detector.system, fit, histogram = pyc.fit_levels(data=fctrace[0:wlength], start_parameters=(0.05, 134900, 20, 0.15, 135100, 20))

    # Plot start stuff
    fig = plt.figure(figsize=(20, 4))
    fig.suptitle(str(wlength))
    plt.subplot2grid((1,4), (0, 0), colspan=3)
    fctrace.plot(0, 10e3, c='blue')
    detector.plot(color='red')
    yl = plt.ylim()

    plt.subplot2grid((1,4), (0, 3))
    histogram.plot(inverted=True, label='Histogram')
    fit.plot(histogram.bins, inverted=True, label='Fit')
    detector.plot(color='red')
    plt.ylim(*yl)
    fig.tight_layout()
  
    for window in fctrace.windows(wlength, start=0):
        #window = pd.rolling_mean(window, avg)[avg:]
            
        # Fit time windowth
        detector.system, fit, histogram = pyc.fit_levels(window, fit.parameters)

        # Digitize data
        detector.digitize(window, signal)

        time.add(signal)

        # Flush everthing
        signal.flush()

    time.plot(ax_t, marker='o', linestyle='-', markersize=2, normed=True, log=True, label=str(wlength))
    ax_t.set_xlim(0, 500)
    ax_t.set_ylim(0.0001, 0.012);
    ax_t.legend()
    fig_t.tight_layout()
    #fig_t.savefig('times_wlength.pdf')
```

```python
%matplotlib inline 

# Create instances
fctrace  = pyc.Trace('data/*.count', datatype='int')

# Determin start parameters
wlength = 1e5


detectors = pyc.MultiDetector(pyc.product_detector(average=[1], nsigma=[0.5, 1, 1.5, 2.0, 2.5]))
times = pyc.MultiTime(pyc.ptime([0,1], factor=len(detectors)))

signals = [pyc.Signal() for i in range(len(detectors))]

detectors.system, fit, histogram = pyc.fit_levels(data=fctrace[0:wlength], start_parameters=(0.05, 134900, 20, 0.15, 135100, 20))

# Plot start stuff
for detector in detectors:
    fig = plt.figure(figsize=(20, 4))
    fig.suptitle(str(detector.nsigma))
    plt.subplot2grid((1,4), (0, 0), colspan=3)
    fctrace.plot(0, 10e3, c='blue')
    detector.plot(color='red')
    yl = plt.ylim()

    plt.subplot2grid((1,4), (0, 3))
    histogram.plot(inverted=True, label='Histogram')
    fit.plot(histogram.bins, inverted=True, label='Fit')
    detector.plot(color='red')
    plt.ylim(*yl)
    fig.tight_layout()
  
for window in fctrace.windows(wlength, start=0):
           
    # Fit time windowth
    detectors.system, fit, histogram = pyc.fit_levels(window, fit.parameters)

    # Digitize data
    #detectors.digitize(window, signals)
    #times.add(signals)
    #signals.flush()
    
    for detector, time, signal in zip(detectors, times, signals):
        detector.digitize(window, signal)

        time.add(signal)

        # Flush everthing
        signal.flush()

fig_t, ax_t = plt.subplots(figsize=(20, 5))
fig_t.suptitle('sigma')
for detector, time in zip(detectors, times):
    time.plot(ax_t, marker='o', linestyle='-', markersize=2, normed=True, log=True, label=str(detector.nsigma))
    ax_t.set_xlim(0, 500)
    ax_t.set_ylim(0.0001, 0.012);
    ax_t.legend()
    fig_t.tight_layout()
#fig_t.savefig('times_sigma.pdf')
```
