```python
trace = pyc.Trace('/run/media/wagner/FREECOM HDD/Bayer Dot/Dg3 0.888 Dg1 0.573 phonon cuppling line/data/*.count', datatype='int')

mdetector = pyc.MultiDetector(pyc.pdetector(nsigma=[0.5, 1, 1.5, 2, 2.5]))
msignal   = pyc.MultiSignal(pyc.psignal(   nr=len(mdetector)))
mtime0    = pyc.MultiTime(  pyc.ptime([0], nr=len(mdetector))) 
mtime1    = pyc.MultiTime(  pyc.ptime([1], nr=len(mdetector)))

fft = pyc.FFT(samplerate=100e3)
wlength = 100e3

mdetector.system, fit, histo = pyc.fit_levels(trace[0:wlength], (0.005, 134950, 20, 0.015, 135075, 20))

for window in trace.windows(wlength, nr=60 * 60):
    
    # FFT Trsansformation
    fft.add(window)
    
    # fit the data and use system
    mdetector.system, fit, histo = pyc.fit_levels(window, fit.parameters)
    
    # Run all detectors an window
    for detector, signal in izip(mdetector, msignal):
        detector.digitize(window, signal)
    
    # Exctract all time from signals
    for time0, time1, signal in izip(mtime0, mtime1, msignal):
        time0.add(signal)
        time1.add(signal)
    
    msignal.flush()
```

# Simulation

```python
# Simulation

# Parameters Signal
rate0 = 1250
level0 = 0
rate1 = 750
level1 = 100
delay = 50

# Parameters Sinus
freq = 8.
noise_sigma = (level1 - level0) / 15.
amp = (level1 - level0) / 15.

strace = pys.BiTrace(rate0=rate0, rate1=rate1, level0=level0, level1=level1, noise_sigma=noise_sigma, delay=delay)
points = int(10 * 100e3)
x = np.linspace(0, points, points)

sin = amp * freq / (2 * np.pi) * np.sin(2 * np.pi / freq * x)

data = strace.range(points) + sin
plt.figure(figsize=(20, 4))
plt.plot(data[:2500])
plt.ylim(-25, 125)
```
