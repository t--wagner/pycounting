# Simulation

```python
# Simulation

# Parameters Signal
rate0 = 1250
level0 = 0
rate1 = 750
level1 = 100
delay = 50

# Parameters Sinus
freq = 8.
noise_sigma = (level1 - level0) / 15.
amp = (level1 - level0) / 15.

strace = pys.BiTrace(rate0=rate0, rate1=rate1, level0=level0, level1=level1, noise_sigma=noise_sigma, delay=delay)
points = int(10 * 10 * 100e3)
x = np.linspace(0, points, points)

sin = amp * freq / (2 * np.pi) * np.sin(2 * np.pi / freq * x)

data = strace.range(points) + sin
plt.figure(figsize=(20, 4))
plt.plot(data[:2500])
plt.ylim(-25, 125)
```

```python
mdetector = pyc.MultiDetector(pyc.pdetector(nsigma=[0.5, 1, 2, 3]))
msignal = [pyc.Signal() for i in range(len(mdetector))]
mtime0 = pyc.MultiTime(pyc.ptime([0], factor=len(mdetector)))
mtime1 = pyc.MultiTime(pyc.ptime([1], factor=len(mdetector)))

mdetector.system, fit, histogram = pyc.fit_levels(data, (0.001, level0, noise_sigma, 0.001, noise_sigma, 10))
```

```python
histogram.plot(inverted=True)
fit.plot(histogram.bins, inverted=True)
mdetector.plot();
```

```python
mdetector.digitize(data, msignal)
mtime0.add(msignal)
mtime1.add(msignal)
for signal in msignal: signal.flush()
```

```python
%matplotlib inline
fig = plt.figure(figsize=(20, 10))
ax = fig.add_subplot(111)

for detector, time0, time1 in zip(mdetector, mtime0, mtime1):
    time0.plot(label=detector.nsigma, marker='.', log=False)
    time1.plot(label=detector.nsigma, marker='.', log=False)

ax.legend()
ax.set_xlim(0, 300)
ax.set_ylim(0, 0.02)
fig.savefig('simulation_sanfter_uebergang.pdf')
```

```python
path = '/run/media/wagner/FREECOM HDD/Counting/Counting/03_g3_-0.620V/trace/*.bin'
trace = pyc.Trace(path, 'ushort')
```

```python
fig = plt.figure(figsize=(20, 4))
trace.plot(0, 10e3)
```

```python
wlength = 500e3

mdetector = pyc.MultiDetector(pyc.pdetector(nsigma=[0.25, 0.5, 1, 2, 3]))
mtime0    = pyc.MultiTime(pyc.ptime([0], factor=len(mdetector)))
mtime1    = pyc.MultiTime(pyc.ptime([1], factor=len(mdetector)))
msignal   = [pyc.Signal() for i in range(len(mdetector))]

mdetector.system, fit, histogram = pyc.fit_levels(trace[:wlength], (0.015, 31770, 17, 0.010, 31940, 17))

histogram.plot()
fit.plot(histogram.bins)

for window in trace.windows(wlength, start=0, nr=5000):
    mdetector.system, fit, histogram = pyc.fit_levels(window, fit.parameters)
    mdetector.digitize(window, msignal)
    mtime0.add(msignal)
    mtime1.add(msignal)
    for signal in msignal:
        signal.flush()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(20, 10))


for detector, time0, time1 in zip(mdetector, mtime0, mtime1):
    line = time0.plot(ax=ax, label=detector.nsigma, marker='.')
    time1.plot(ax=ax, label=detector.nsigma, color=line.get_color(), marker='.')

ax.set_xlim(0, 200)
ax.set_ylim(1e-3, 2e-2)
ax.legend()

#fig.savefig('altes_Signal.pdf')
```
